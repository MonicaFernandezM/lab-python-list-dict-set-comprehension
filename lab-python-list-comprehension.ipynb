````python
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
````

````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter te quantity for {product}:"))
                 for product in products}
    return inventory
````

````python
def get_customer_orders():
    n_of_order = int(input("Enter the number of customer orders: "))

    customer_orders = {
        input("Enter the name of a product: ")
        for _ in range(n_of_order)
    }
    return customer_orders
````

````python
def calculate_total_price(customer_orders):
    prices =[float(input(f"Enter the price of {product}: ")) for product in customer_orders]
    total = sum(prices)
    return total
````

````python
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1

    #Reduce products with zero
    inventory = {product:quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory
````

````python
def calculate_order_statistics(customer_orders, products):
    total_ordered = len(customer_orders)
    percentage_order = (total_ordered / len(products)) *100
    order_statistics = (total_ordered, percentage_order)
    return total_ordered, percentage_order, order_statistics
````

````python
def print_order_statistics(order_statistics):
    print("Order Statistics: ")
    print(f"Total Products Ordered: {order_statistics[0]}")
    print(f"Percentage of Products Ordered: {order_statistics[1]}%")
````

````python
def print_updated_inventory(inventory):
    print("Updated Inventary: ")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
````

````python
inventory = initialize_inventory(products)
````

````python
customer_orders = get_customer_orders()
````

````python
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_sta = print_order_statistics(order_statistics)
print_updated_inv = print_updated_inventory(inventory)

total_price = calculate_total_price(customer_orders)
print(f"Total Price: {total_price}")
````
